LINEAR REGRESSION MODEL-

Separating the columns into features and target-

```python
features=df.drop('CGPA',axis=1)
target=df['CGPA']
```

Scaling the data using MinMax scaler-

```python
from sklearn.preprocessing import MinMaxScaler
mns=MinMaxScaler()
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
for i in range(0,100):
    features_train,features_test,target_train,target_test=train_test_split(features,target,test_size=0.2,random_state=i)
    lr.fit(features_train,target_train)
    pred_train=lr.predict(features_train)
    pred_test=lr.predict(features_test)
    print(f'At random state {i},the training accuracy is:- {r2_score(target_train,pred_train)}')
    print(f'At random state {i},the testing accuracy is:- {r2_score(target_test,pred_test)}')
    print('\n')
```

```python
features_train,features_test,target_train,target_test=train_test_split(features,target,test_size=0.2,random_state=5)
```

We choose random state as 5 because the difference between training and testing accuracy is the least there.

```python
lr.fit(features_train,target_train)
```

```python
pred_test=lr.predict(features_test)
```

```python
print(r2_score(target_test,pred_test))
```

Cross validation of the model-

```python
Train_accuracy=r2_score(target_train,pred_train)
Test_accuracy=r2_score(target_test,pred_test)
from sklearn.model_selection import cross_val_score
for j in range(2,10):
    cv_score=cross_val_score(lr,features,target,cv=j)
    cv_mean=cv_score.mean()
    print(f'At cross fold {j} the cv score is {cv_mean} and accuracy score for training is {Train_accuracy} and accuracy for the testing is {Test_accuracy}')
    print('\n')
```

Here, we have handled the problem of the overfitting and the underfitting by checking the training and testing score.

```python
plt.figure(figsize=(8,6))
plt.scatter(x=target_test,y=pred_test,color='r')
plt.plot(target_test,target_test,color='b')
plt.xlabel('Actual CGPA',fontsize=14)
plt.ylabel('Predicted CGPA',fontsize=14)
plt.title('Linear Regression',fontsize=14)
plt.savefig('lr.png')
plt.show()
```

Best fit line is covering most of the datapoints which shows good fit of our model.

Regularization-

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
parameters={'alpha': [.0001, .001, .01, .1, 1, 10],
           'random_state':list(range(0,10))}
ls=Lasso()
clf= GridSearchCV(ls,parameters)
clf.fit(features_train,target_train)
print(clf.best_params_)
```

Final model training-

```python
ls=Lasso(alpha=0.01,random_state=0)
ls.fit(features_test,target_test)
ls_score_testing=ls.score(features_test,target_test)
pred_ls=ls.predict(features_test)
ls_score_testing*100
```

Ensemble technique-

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
parameters={'criterion': ['mse','mae'],
           'max_features':['auto','sqrt','log2']}
rf=RandomForestRegressor()
clf= GridSearchCV(rf,parameters)
clf.fit(features_train,target_train)
print(clf.best_params_)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
rf=RandomForestRegressor(criterion='mse',max_features='sqrt')
rf.fit(features_train,target_train)
rf.score(features_train,target_train)
pred_decision= rf.predict(features_test)
rfs=r2_score(target_test,pred_decision)
print('R2 Score:',rfs*100)
rfscore=cross_val_score(rf,features,target,cv=5)
rfc= rfscore.mean()
print('Cross val score:',rfc*100)
```

We are getting model accuracy and cross validation both above 94% which shows our model is performing very good.

We choose cv=5 here because, the accuracy score there is the highest.
